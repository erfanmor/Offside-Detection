# Offside Class Documantion
## Import libraries

```python
from Offside import OffsideClass
import os
```

## Define a object from OffsideClass
For create object we should pass the image path to class

```python
path = os.path.join("./images/offside.jpg")
offside = OffsideClass(path)
```

## Lets show and explain object features

### 1. **static_line_detection**  method
- This method use to select parallel lines (two) in image and return a list of two lines [[x1,y1, x2,y2], [x3,y3, x4,y4]]
- We can pass the max_width attribute for setting window size (default 800).
####
- Guide:
    - Use **ESC** button to exit and select line
    - Use **'c'** button to clean image
    - Ues **right-click** to delete last selected point

```python
offside.static_line_detection(max_width=700)
```

### 2. **auto_line_detection**  method
- This method use to select parallel lines (two) in image and return a list of two lines [[x1,y1, x2,y2], [x3,y3, x4,y4]]
- This method use from trackbars to easy line detection.
- We can pass the max_width attribute for setting window size (default 500)
####
- Trackbars Guide:
    - The lower limit of line detection
    - The higher limit of line detection
    - angle is lines adge range
    - for save final image use save trackbar
    - for Onf and Off use the last trackbar
- Deleting extra lines Guide:
    - Enter number of line where you want to delete it in input box and click on Add.
    - To see final image (before delete lines) off and on again power trackbar and then save it or your sure from final image you can save it.
####
- Press any key to clone line detection window.
- After your detection is done make sure about all windows are closed.

```python
offside.auto_line_detection(max_width=500)
```

### 3. **find_vanishing_point**  method
- This method is used to find vanishing point.
- For attributes get **line1** and **line2** (two parallel lines).
- Return **vanishing point**.

```python
line1, line2 = [[382, 307, 384, 149], [224, 292, 290, 142]]
offside.find_vanishing_point(line1, line2)
```

### 4. **draw_lines**  method
- This method get vanishing point and draw offside lines (max two lines)
- For attributes get **vanishing_point** and **max_width** to set window size (defualt 800)
####
- Guide:
    - Use **right-click** to draw line.
    - **Left-click** once to change the line color.
    - Press **'s'** button on keyboard to save image as **'final.jpg'**
    - Press **ESC** button to exit.

```python
van_point = (440, -78)
offside.draw_line(vanishing_point=van_point)
```

### 5. **run_as** method
- To fast run full code for detection and draw lines use it.
- Get type as attribute (its str and is one of the **'static'** and **'auto'**)
####
#### Execute this code to rus **static** mode:
- First select two parallels line and then draw offside lines

```python
offside.run_as('static')
```

### 6. **increase_contrast** method
- This method use to create higher contrast image.
- High contrast image save in equalized_image value
- For attributes get **show** if True you can see final image and compare with original image and if false image only save and you cant see it directly.

This method help you to have better image preprocessing

```python
offside.increase_contrast(show=True)
offside.equalized_image.shape
```

### 7. **player_detection** method
- Tihs method use from pretrained model (yolov8n) to detect humans and other objects in image (player and ball)
- Get **show** as attribute and if its True you can see detection image (detected players)

```python
offside.player_detection(show=True)
```
